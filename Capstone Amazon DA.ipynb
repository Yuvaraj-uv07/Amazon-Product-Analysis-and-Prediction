```python
# reading the csv file

df = pd.read_csv("amazon_fashion.csv")
```

```python
# checking for duplicate values

df.duplicated().sum()
```

```python
# dropping the duplicate values

df = df.drop_duplicates()
```

```python
# Checking for null values

df.isnull().sum()
```

```python
# Checking for rows which have all rows as null

df[df['Product Name'].isnull()]
```

```python
# Dropping the rows which have all rows as null

df = df.dropna(subset=['Product Name'])
```

```python
# Cheking null values in 'Price' Column

df[df['Price'].isnull()]
```

```python
# Dropping it because price related columns are null

df = df.dropna(subset=['Price'])
```

```python
# Filling the 'Rating' column null value with 0

df['Rating'] = df["Rating"].fillna(0)
```

```python
# Taking the brand alone Only apply to rows where Category is "Men's Watches"

df.loc[df['Category'] == "Men's Watches", 'Product Name'] = df['Product Name'].str.split().str[0]
```

```python
# Renaming the columns

df = df.rename(columns={
    'Category': 'Product',
    'Product Name': 'Brand'
})
```

```python
# Creating new column with the category

def extract_gender(cat):
    cat = cat.lower()
    if cat.startswith("men"):
        return "Men"
    elif cat.startswith("women"):
        return "Women"
    elif cat.startswith("kids"):
        return "Kids"
    else:
        return "Other"

df['Category'] = df['Product'].apply(extract_gender)
```

```python
df["Sub Category"] = df["Product"].str.split().str[1]
# Remove trailing 's' from category names
df['Sub Category'] = df['Sub Category'].str.rstrip('s')
```

```python
df.reset_index()
```

```python
df.dtypes
```

```python
for col in ['Price', 'Original Price', 'Discount %', 'Discount Amount']:
    is_true_float = (df[col] % 1 != 0).any()
    print(f"{col} has true floats:", is_true_float)
```

```python
# changing data type into int

df[['Price','Discount Amount']] = df[['Price','Discount Amount']].astype(int)
```

```python
# Define bins
bins = [151, 6000, 12000, 18000, 24000]

# Labels reflecting expensiveness
labels = ['Budget', 'Mid-Range', 'Expensive', 'Luxury']

# Apply binning
df['price_range'] = pd.cut(df['Price'], bins=bins, labels=labels, right=False)

# Preview
df[['Price', 'price_range']].head()
```

```python
# Example: Replace specific product names
df['Sub Category'] = df['Sub Category'].replace({
    'Dresse': 'Dress',
    'Watche': 'Watch'})
```

```python
df = df[['Category', 'Sub Category', 'Brand', 'Product',
         'Price', 'Original Price', 'Discount %', 'Discount Amount',
         'price_range', 'Availability', 'Rating', 'Is Prime',
         'Product URL', 'Image URL', 'Date Scraped']]
```

```python
import mysql.connector



# MySQL config
config = {
    'host': 'localhost',     # Change if using remote host
    'user': 'root',
    'password': 'root',
    'database': 'Amazon'
}

try:
    conn = mysql.connector.connect(**config)
    if conn.is_connected():
        print("✅ Connected to MySQL!")

    cursor = conn.cursor()

    insert_query = """INSERT INTO Amazon_Data 
(Category, Sub_Category, Brand, Product, Price, Original_Price, Discount_Per, Discount_Amount, Price_Range, Availability, Rating, Is_Prime, Product_URL, Image_URL, Date_Scrapped)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s,%s,%s)
"""

  
    for _, row in df.iterrows():
        cursor.execute(insert_query, (
            row['Category'], row['Sub_Category'], row['Brand'], row['Product'],row['Price'],
            row['Original_Price'], row['Discount_Per'],row['Discount_Amount'], row['Price_Range'], row['Availability'],
            row['Rating'], row['Is_Prime'], row['Product_URL'], row['Image_URL'], row['Date_Scrapped']
        ))

    conn.commit()
    print(" Data inserted successfully!")

except mysql.connector.Error as err:
    print(f" Error: {err}")

finally:
    try:
        if conn.is_connected():
            cursor.close()
            conn.close()
            print(" MySQL connection closed.")
    except NameError:
        print("ℹ No active MySQL connection to close.")
```

```python
numeric_cols = df_scaled.select_dtypes(include=[np.number]).columns

# Apply Yeo-Johnson transformation
scaler = PowerTransformer(method='yeo-johnson')
df_scaled = df_scaled.copy()  # Make a copy to store transformed values
df_scaled[numeric_cols] = scaler.fit_transform(df_scaled[numeric_cols])
```

```python
for column in df_scaled.select_dtypes(include='number').columns:
    stat, p_value = shapiro(df_scaled[column])
    print(f'Column: {column}')
    print(f'Test Statistic: {stat}')
    print(f'P-Value: {p_value}')
    
    # Check if the column is normally distributed
    if p_value > 0.05:
        print(f'The column "{column}" is normally distributed.\n')
    else:
        print(f'The column "{column}" is not normally distributed.\n')
```

```python
for column in df_scaled.select_dtypes(include='number').columns:
    if column != 'Price': 
        stat, p_value = levene(df_scaled['Price'], df_scaled[column])
        print(f'Comparing Price with Column: {column}')
        print(f'Test Statistic: {stat}')
        print(f'P-Value: {p_value}')
    
        if p_value > 0.05:
            print(f'The variances between "Price" and "{column}" are equal (p-value > 0.05).\n')
        else:
            print(f'The variances between "Price" and "{column}" are not equal (p-value <= 0.05).\n')
```

```python
# Step 3: Use Elbow method to find optimal k
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method to Determine Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
# Fit KMeans with optimal k (example: k=7)
optimal_k = 7  # replace with the elbow point you identify visually
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['cluster'] = kmeans.fit_predict(df_scaled)

# Numeric summary by cluster
numeric_columns = df.select_dtypes(include=['number']).columns
cluster_summary = df.groupby('cluster')[numeric_columns].mean().round(2)
print("Numerical Summary by Cluster:")
print(cluster_summary)
```

```python
# Most common category in each cluster (optional)
categorical_columns = df.select_dtypes(include='object').columns

for col in categorical_columns:
    print(f"\nMost common '{col}' in each cluster:")
    print(df.groupby('cluster')[col].agg(lambda x: x.mode().iloc[0]))
```

<p style="background-color:#232F3E; font-family:calibri; color:#FF9900; padding:10px; font-size:150%; text-align:center; border-radius:15px 50px;">
XG Boost</p>

```python
# Convert to pandas DataFrame / Series
x_train = pd.DataFrame(x_train)
y_train = pd.Series(y_train)

#  Split labeled and unlabeled data correctly
x_label, y_label = x_train.iloc[:686], y_train.iloc[:686]
x_unlabelled = x_train.iloc[686:]

print(f"x_label : {x_label.shape} \ny_label : {y_label.shape} \nx_unlabelled : {x_unlabelled.shape}")

#  Number of classes
n_classes = len(y_label.unique())

#  Define model
model = xgb.XGBClassifier(
    objective='multi:softmax',
    num_class=n_classes,
    use_label_encoder=False,
    eval_metric='mlogloss',
    random_state=42
)

#  Initial training
model.fit(x_label, y_label)

#  Semi-supervised loop
while True:
    y_pred_proba = model.predict_proba(x_unlabelled)

    # Get high-confidence indexes
    index = [i for i, prob in enumerate(np.max(y_pred_proba, axis=1)) if prob > 0.90]

    if len(index) == 0:
        print("No more confident samples found.")
        break

    # Get confident samples and predictions
    confident_x = x_unlabelled.iloc[index]
    confident_y = model.predict(confident_x)

    # Drop confident samples from unlabeled set
    x_unlabelled = x_unlabelled.drop(x_unlabelled.index[index]).reset_index(drop=True)

    # Append confident samples to labeled dataset
    x_label = pd.concat([x_label, confident_x], ignore_index=True)
    y_label = pd.concat([y_label, pd.Series(confident_y)], ignore_index=True)

    print(f" Added {len(index)} confident samples. Total labeled: {x_label.shape[0]}")

#  Retrain on the full labeled data
model.fit(x_label, y_label)

#  Predict and evaluate
y_pred_xg = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred_xg)
print(f"\n Final Model Accuracy: {accuracy * 100:.2f}%")
```
